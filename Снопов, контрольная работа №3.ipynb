```python
from sympy import *
```

```python
x,y,a,b = symbols('x,y,a,b', real = True)
point = (3*sqrt(2), -4)
# F(c,0) - focus, -> 2c - distance; c = sqrt(a**2 + b**2) = 5; a**2 + b**2 = 25
solut = solve([point[0]**2/a**2 - point[1]**2/b**2 -1, a**2 + b**2 - 25], [a,b], domain = S.Reals)
print(solut)
```

```python
#Находим только те a,b, которые больше 0(по геом. соображениям)
solution = []
for s in solut:
    Positive = True
    for p in s:
        if p<0:
            Posiitve = False
    if Positive:
        solution = s

print(solution)
```

```python
a = solution[0]
b = solution[1]
plot_implicit(x**2/a**2 - y**2/b**2 - 1, (x, -10,10), (y,-10,10))
```

#### Номер 2

```python
#2*a1 + 5*d = 3/2 ; (a1+d)*(a1+4*d) = 1/2
a1,d,n = symbols('a1, d,n')
solut = solve([2*a1 + 5*d - 3/2, (a1+d)*(a1+4*d) - 1/2])
print(solut)
```

```python
#Первая последовательность
a,d = list(solut[0].values())
first = SeqFormula(a + (n-1)*d, (n,1, 12))
print(list(map(lambda x: round(x,3), first[:])))  #округление до 3 знака после запятой
```

```python
#Вторая последовательность
a,d = list(solut[1].values())
first = SeqFormula(a + (n-1)*d, (n,1, 12))
print(list(map(lambda x: round(x,3), first[:])))  #округление до 3 знака после запятой
```

#### Номер 3

```python
def diff(roots):
    return max(roots) - min(roots)  #разность между максимальным и минимальным
```

```python
x = Symbol('x', real = True)
equat = expand((x-2)**2 * (x+4) * (x**2+2)**2)
roots = solve(equat, domain = S.Reals)
d = diff(roots)
print(roots)
print(d)
```
